```python
v3 = {}

for i in open('genome_Bastian_Greshake_v3_Full_20171225115432.txt'):
    if i[0] != "#":
        ia = i.strip().split('\t')
        v3[ia[0]] = ia[-1]
        
v5 = {}

for i in open('genome_Bastian_GreshakeTzovaras_v5_Full_20171225115452.txt'):
    if i[0] != "#":
        ia = i.strip().split('\t')
        v5[ia[0]] = ia[-1]
        
print('number of snps v3: {}'.format(len(v3)))
print('number of snps v5: {}'.format(len(v5)))
        
identical = 0
different = 0

different_dict = {}

for rs,genotype in v5.items():
    if rs in v3.keys():
        if genotype == v3[rs]:
            identical += 1
        elif len(genotype) == 2:
            if genotype[1] + genotype[0] == v3[rs]:
                identical += 1
            else:
                different += 1
                diff = {'v3': v3[rs], 'v5': genotype}
                different_dict[rs] = diff
        else:
            different += 1
            diff = {'v3': v3[rs], 'v5': genotype}
            different_dict[rs] = diff
print('concordant snps: {}'.format(identical))
print('discordant snps: {}'.format(different))

import myvariant
mv = myvariant.MyVariantInfo()

print('RsID\t        v3\tv5\tphenotype\tpvalue\tpmid\tpublication_year')
for rs in different_dict:
    if different_dict[rs]['v3'] != '--' and different_dict[rs]['v5'] != '--':
        variant = mv.query('dbsnp.rsid:{}'.format(rs),fields='grasp')
        try:
            for pub in variant['hits'][0]['grasp']['publication']:
                pubdate = pub['date_pub']
                pmid = pub['pmid']
                pvalue = pub['p_value']
                phenotype = pub['phenotype']
                print('{}\t{}\t{}\t{}\t{}\t{}\t{}'.format(rs, 
                                                      different_dict[rs]['v3'],
                                                      different_dict[rs]['v5'],
                                                      phenotype, 
                                                      pvalue,
                                                      pmid,
                                                      pubdate[-4:]))
        except:
              print('{}\t{}\t{}\t{}\t{}\t{}\t{}'.format(rs, 
                                                      different_dict[rs]['v3'],
                                                      different_dict[rs]['v5'],
                                                      '-', 
                                                      '-',
                                                      '-',
                                                      '-'))
```
